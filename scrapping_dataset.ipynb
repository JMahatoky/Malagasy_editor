```python
import requests
import os
import time
import re

URL = "https://mg.wikipedia.org/w/api.php"

HEADERS = {
    "User-Agent": "JMahatoky/Malagasy_editor (mahatoky04@gmail.com)"
}

SAVE_PATH = "dataset/raw"
os.makedirs(SAVE_PATH, exist_ok=True)

visited_titles = set()


def sanitize_filename(name):
    # enlever caractères interdits Windows
    name = re.sub(r'[<>:"/\\|?*]', '', name)
    # remplacer espaces par _
    name = name.replace(" ", "_")
    return name


def save_text(title, text):
    safe_title = sanitize_filename(title)
    filename = safe_title + ".txt"
    filepath = os.path.join(SAVE_PATH, filename)

    with open(filepath, "w", encoding="utf-8") as f:
        f.write(text)


def clean_text(text):
    text = text.lower()
    text = re.sub(r"\n+", " ", text)
    text = re.sub(r"[^a-zàâéèêîôû\s]", "", text)
    return text.strip()


def get_random_titles(n=10):
    params = {
        "action": "query",
        "list": "random",
        "rnlimit": n,
        "format": "json"
    }
    res = requests.get(URL, params=params, headers=HEADERS).json()
    return [item["title"] for item in res["query"]["random"]]


def get_page_text(title):
    params = {
        "action": "query",
        "format": "json",
        "prop": "extracts",
        "titles": title,
        "explaintext": True
    }
    res = requests.get(URL, params=params, headers=HEADERS).json()
    pages = res["query"]["pages"]
    return list(pages.values())[0].get("extract", "")


def collect_data(iterations=50):
    for i in range(iterations):
        print(f"Iteration {i+1}")

        titles = get_random_titles(5)

        for title in titles:
            if title in visited_titles:
                continue

            visited_titles.add(title)

            text = get_page_text(title)

            if len(text) < 200:
                continue

            clean = clean_text(text)

            save_text(title, clean)

            print(f"Saved: {title}")

            time.sleep(1)


if __name__ == "__main__":
    collect_data(100)
```

```python
import collections
import random

class MarkovAutocomplete:
    def __init__(self):
        # On stocke : {mot_actuel: {mot_suivant: occurence}}
        self.model = collections.defaultdict(lambda: collections.Counter())

    def train(self, file_path):
        with open(file_path, 'r', encoding='utf-8') as f:
            for line in f:
                # Nettoyage et découpage en mots
                words = line.lower().strip().split()
                # On crée des paires de mots (Bigrams)
                for i in range(len(words) - 1):
                    current_word = words[i]
                    next_word = words[i+1]
                    self.model[current_word][next_word] += 1
        print("Entraînement terminé.")

    def predict(self, current_word, limit=3):
        current_word = current_word.lower()
        if current_word not in self.model:
            return []
        
        # On récupère les mots suivants les plus fréquents
        next_words = self.model[current_word].most_common(limit)
        return [word for word, count in next_words]

# --- Test ---
predictor = MarkovAutocomplete()
predictor.train('data.txt')

# Simulation : l'utilisateur a tapé "Mianatra"
word = "dada"
suggestions = predictor.predict(word)
print(f"Après '{word}', les malagasy disent souvent : {suggestions}")
```

```python
import json
import collections
import re

def train_ngram(file_path, model_path, n=3):
    # Dictionnaire de fréquences : { "mot1 mot2": {"mot3": 5, "mot4": 2} }
    model = collections.defaultdict(lambda: collections.Counter())

    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            # Nettoyage simple : on garde les lettres et les espaces
            line = line.lower().strip()
            words = re.findall(r'\b\w+\b', line)
            
            if len(words) < n:
                continue
                
            # Création des séquences
            for i in range(len(words) - n + 1):
                context = " ".join(words[i : i + n - 1])
                target = words[i + n - 1]
                model[context][target] += 1

    # Convertir en dictionnaire standard pour le JSON
    output_model = {ctx: dict(counts) for ctx, counts in model.items()}
    
    with open(model_path, 'w', encoding='utf-8') as f:
        json.dump({"n": n, "data": output_model}, f, ensure_ascii=False, indent=4)
    
    print(f"Modèle {n}-gram enregistré dans {model_path}")

# Exemple : Entraînement avec des Trigrams (n=3)
train_ngram('data.txt', 'malagasy_model.json', n=3)
```

```python
import json

def get_suggestions(input_text, model_path, nb_suggestions=3):
    with open(model_path, 'r', encoding='utf-8') as f:
        model_data = json.load(f)
    
    n = model_data['n']
    model = model_data['data']
    
    # On nettoie l'entrée et on prend les (n-1) derniers mots
    words = input_text.lower().strip().split()
    context = " ".join(words[-(n-1):])
    
    if context in model:
        # Trier les mots par fréquence
        suggestions = sorted(model[context].items(), key=lambda x: x[1], reverse=True)
        return [word for word, count in suggestions[:nb_suggestions]]
    else:
        return []

# Test de prédiction
phrase_partielle = "eto" # Si n=3, il cherche ce qui suit ces 2 mots
resultats = get_suggestions(phrase_partielle, 'malagasy_global_model.json')
print(f"Suggestions pour '{phrase_partielle}': {resultats}")
```

```python
import json
import collections
import re
import os

def train_ngram_multi(file_list, model_path, n=3):
    # Le dictionnaire qui centralise tout
    model = collections.defaultdict(lambda: collections.Counter())

    for file_path in file_list:
        if not os.path.exists(file_path):
            print(f"⚠️ Attention : Le fichier {file_path} est introuvable. Ignoré.")
            continue
            
        print(f"Lecture de {file_path}...")
        with open(file_path, 'r', encoding='utf-8') as f:
            for line in f:
                # Nettoyage et extraction des mots
                line = line.lower().strip()
                words = re.findall(r'\b\w+\b', line)
                
                if len(words) < n:
                    continue
                
                # Génération des N-grams
                for i in range(len(words) - n + 1):
                    context = " ".join(words[i : i + n - 1])
                    target = words[i + n - 1]
                    model[context][target] += 1

    # Conversion pour sauvegarde JSON
    output_model = {ctx: dict(counts) for ctx, counts in model.items()}
    
    with open(model_path, 'w', encoding='utf-8') as f:
        json.dump({"n": n, "data": output_model}, f, ensure_ascii=False, indent=4)
    
    print(f"✅ Entraînement terminé ! Modèle {n}-gram enregistré dans {model_path}")

# --- Exécution ---
fichiers_entrainement = ['data.txt', 'journal.txt']
train_ngram_multi(fichiers_entrainement, 'malagasy_global_model.json', n=3)
```
